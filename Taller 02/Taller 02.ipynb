```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return x**3 - 3*x**2 + x - 1

# Generate x values
x = np.linspace(-2, 4, 400)

# Generate y values
y = f(x)

# Create the plot
plt.plot(x, y, label='y = x^3 - 3x^2 + x - 1')

# Add title and labels
plt.title('Plot of the equation y = x^3 - 3x^2 + x - 1')
plt.xlabel('x')
plt.ylabel('y')

# Add a grid
plt.grid(True)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
