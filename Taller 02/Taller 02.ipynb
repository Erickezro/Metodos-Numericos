```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# Define the function f(x) and its derivative f'(x)
def f(x):
    return x**3 - 3*x**2 + x - 1

def df(x):
    return 3*x**2 - 6*x + 1

# List to store explored points
points = []

# Wrapper function to track points
def f_with_tracking(x):
    points.append(x)
    return f(x)

# Initial value and finding root using Newton's method
x0 = 100
x_r = newton(f_with_tracking, x0, fprime=df)

# Plotting the trajectory of the roots
plt.figure(figsize=(10, 6))
plt.plot(points, 'ro-', label="Trajectory of Approximations")
plt.xlabel("Iteration")
plt.ylabel("x-value")
plt.title("Trajectory of Newton-Raphson Approximations")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Definir la función y su derivada
def y(x):
    return x**3 - 3*x**2 + x - 1

def dy_dx(x):
    return 3*x**2 - 6*x + 1

# Implementar el método de Newton-Raphson que almacena las estimaciones
def newton_raphson_trace(func, deriv, x0, tol=1e-6, max_iter=50):
    x = x0
    estimaciones = [x]  # Lista para almacenar las estimaciones
    for i in range(max_iter):
        fx = func(x)
        dfx = deriv(x)
        if abs(dfx) < 1e-10:
            print("La derivada es cero, el método falla.")
            return None, estimaciones
        x_new = x - fx / dfx
        estimaciones.append(x_new)  # Guardar la nueva estimación
        if abs(x_new - x) < tol:
            return x_new, estimaciones
        x = x_new
    print("Máximo de iteraciones alcanzado.")
    return x, estimaciones

# Definir tres estimaciones iniciales diferentes
initial_guesses = [1.5, 2, 2.5]

# Crear la gráfica
plt.figure(figsize=(10, 6))

for guess in initial_guesses:
    root, trace = newton_raphson_trace(y, dy_dx, guess)
    plt.plot(trace, marker='o', label=f'Estimación inicial = {guess}')

plt.xlabel('Iteración')
plt.ylabel('Valor de x')
plt.title('Trayectoria de las raíces usando el método de Newton-Raphson')
plt.legend()
plt.grid(True)
plt.show()
```
