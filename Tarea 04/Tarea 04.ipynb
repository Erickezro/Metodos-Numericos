### Método de bisección

```python

# Método de bisección
def bisection_method(a, b, tol):
    if f(a) * f(b) >= 0:
        print("No se puede aplicar el método de bisección.")
        return None
    
    iteration = 1
    while (b - a) >= tol:
        c = (a + b) / 2
        print(f"Iteración {iteration}: a={a:.6f}, b={b:.6f}, c={c:.6f}, f(c)={f(c):.6f}")
        
        if f(c) == 0:
            return c
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iteration += 1
    
    return (a + b) / 2
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función
def f(x):
    return x**3 - x - 1

# Método de bisección
def bisection_method(a, b, tol):
    if f(a) * f(b) >= 0:
        print("No se puede aplicar el método de bisección.")
        return None
    
    iterations = []
    iteration = 1
    while (b - a) >= tol:
        c = (a + b) / 2
        iterations.append((a, b, c, f(c)))
        
        print(f"Iteración {iteration}: a={a:.6f}, b={b:.6f}, c={c:.6f}, f(c)={f(c):.6f}")
        
        if f(c) == 0:
            return c
        elif f(c) * f(a) < 0:
            b = c
        else:
            a = c
        iteration += 1
    
    return (a + b) / 2, iterations

# Definimos los límites y la tolerancia
a = 1
b = 2
tol = 1e-6

# Ejecutamos el método de bisección
root, iterations = bisection_method(a, b, tol)

# Graficamos la función y las aproximaciones
x = np.linspace(0.5, 2.5, 400)  # Rango para la gráfica
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = x^2 - x - 1', color='blue')
plt.axhline(0, color='black', lw=0.5, ls='--')  # Línea y=0
plt.axvline(root, color='red', linestyle='--', label=f'Raíz aproximada: x = {root:.6f}')

# Graficamos los puntos de las iteraciones
for a_val, b_val, c_val, f_c in iterations:
    plt.scatter(c_val, f_c, color='orange')  # Punto c
    plt.scatter(a_val, f(a_val), color='green')  # Punto a
    plt.scatter(b_val, f(b_val), color='purple')  # Punto b

plt.title('Método de Bisección')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.xlim(0, 2.5)
plt.ylim(-2, 2)
plt.show()

print(f"\nRaíz encontrada: {root:.6f}")
```
