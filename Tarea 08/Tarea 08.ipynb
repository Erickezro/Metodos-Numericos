## 1.
| \($x_i\$) | 4.0 | 4.2 | 4.5 | 4.7 | 5.1 | 5.5 | 5.9 | 6.3 | 6.8 | 7.1 |
|---------|------|------|------|------|------|------|------|------|------|------|
| \(y_i\) | 102.56 | 130.11 | 113.18 | 142.05 | 167.53 | 195.14 | 224.87 | 256.73 | 299.50 | 326.72 |


```python
import numpy as np

# Datos
xi = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
yi = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])
```

```python
# Funcione para calcular la Suma de errores al cuadrado y el error cuadrático medio (ECM)

#SSE (Suma de errores al cuadrado SEC): Si buscas el error absoluto total del modelo.
def error_sec(yi, y_hat):
    return np.sum((yi - y_hat)**2)

#ECM (Error cuadrático medio): Si quieres el promedio del error, que permite comparar modelos de diferentes tamaños de datos.
def error_ecm(yi, y_hat):
    return np.mean((yi - y_hat)**2)
```

### a.

```python
import matplotlib.pyplot as plt

# Ajuste polinómico de grado 1 (lineal)
coef1 = np.polyfit(xi, yi, 1)  # Coeficientes del polinomio
y_pred1 = np.polyval(coef1, xi)  # Predicción de valores y_hat
err_sec1 = error_sec(yi, y_pred1)  # Cálculo de SEC
err_ecm1 = error_ecm(yi, y_pred1)  # Cálculo del error ECM

# Resultados
print("Ajuste lineal (Polinomio de grado 1):")
print(f"   Ecuación: y = {coef1[0]:.4f}x + {coef1[1]:.4f}")
print(f"   Suma de errores al cuadrado (SEC): {err_sec1:.4f}")
print(f"   Error cuadrático medio (ECM): {err_ecm1:.4f}\n")

# Visualización del ajuste
plt.scatter(xi, yi, color='blue', label='Datos reales')
plt.plot(xi, y_pred1, color='red', label='Ajuste lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste lineal por mínimos cuadrados')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos proporcionados
act_puntos = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 27, 29, 21, 28, 28, 26, 30, 24])
prome_puntos = np.array([3.84,3.21,3.23,3.63,3.75,3.20,3.41,3.38,3.53,2.03, 3.75,3.65,3.87,3.75,1.66,3.12,2.96,2.92,3.10,2.81])

# Graficar los puntos de datos
plt.figure(figsize=(8, 5))
plt.scatter(act_puntos, prome_puntos, color='blue', label='Datos reales')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.title('Relación entre la puntuación ACT y el promedio de puntos')
plt.grid(True)

# Ajuste por mínimos cuadrados (recta de mejor ajuste)(x,y, #)
coef3 = np.polyfit(act_puntos, prome_puntos, 1)
y_pred = np.polyval(coef3, act_puntos)

# Graficar la recta de ajuste
plt.plot(act_puntos, y_pred, color='red', label=f'Ajuste por mínimos cuadrados: y = {coef3[0]:.4f}x + {coef3[1]:.4f}')

# Mostrar leyenda y la gráfica
plt.legend()
plt.show()

# Mostrar la ecuación de la recta
print(f'Ecuación de la recta de ajuste: y = {coef3[0]:.4f}x + {coef3[1]:.4f}')
```
